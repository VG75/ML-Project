So lets see some visualisation of this so that we might get a better idea into this data set

```python
import seaborn as sns
```

```python
result = df.groupby('State').sum()
result
```

```python
expenditure = ['R&D Spend','Administration','Marketing Spend']
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
result= result.reset_index()
result
```

```python
melt_df = pd.melt(result, id_vars='State', value_vars=result.columns[1:4])
```

```python

sns.barplot(data=melt_df, x='State', y='value', hue='variable', hue_order=expenditure)
plt.title('Spending by differnt states')
plt.legend(title='Spending Category')
plt.show()
```

From the above graph ceratin inferences can be drawn like,
State wise insight like what is the actual spending of each state on each category, what's they spend on the most and where ther is improvement and we can also see which states spend maximum and minimum on each category in comparison to other.

```python
plt.pie(result['Profit'], labels=result['State'], autopct='%1.1f%%')
plt.title('Profit Distribution')
plt.show()
```

It is worthwhile to note that even though the expenditure of each state is so diverse the profits are almost similar. So this might prove to be one of the useful conclusion for further analysis on this data.

Since this data contains limited amount of data and only profits and spending making various conclusions becomes difficult as we dont know actually why these results are occuring.
